```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv').drop(['ID','Doors'], axis=1)
df['Leather interior'].unique
```

```python
def YNToNum(choice):
    choice = choice.lower()
    if choice == 'yes':
        return 1
    elif choice == 'no':
        return 0
df['Leather interior'] = df['Leather interior'].apply(lambda x: YNToNum(x))
```

```python
df.Mileage = df.Mileage.apply(lambda x: float(int(x.split()[0]))/1000)
df.head()
```

```python
df['Cylinders'] = df['Cylinders'].apply(lambda x: int(x))
df
```

```python
df = df.drop(labels='Levy', axis='columns')
df
```

```python
def DriveWheelsConv(n):
    n = n.lower()
    if n == '4x4':
        return 3
    elif n == 'rear':
        return 2
    elif n == 'front':
        return 1
df['Drive wheels'] = df['Drive wheels'].apply(lambda x: DriveWheelsConv(x))
```

# Reducing Feature size:

```python

def OtherColor(c, color_list = ['purple', 'yellow', 'pink']):
    if str(c).lower() in color_list:
        return 'other'
    else:
        return str(c).lower()
df['Color'] = df.Color.apply(lambda x: OtherColor(x))
df['Color'].value_counts().plot.bar()
#df['Color'].value_counts()
df['Color'].unique()
```

# Cleaning Up Engine Volume
(too many categories due to 'TURBO')

```python
#Getting raw number 
def convToVol(n):
    n = str(n).strip()
    tokens = n.split()
    return float(tokens[0])
#Splitting Turbo and non turbo engine types
def TurboClass(n):
    n = str(n).strip()
    tokens = n.split()
    if len(tokens) == 1:
        return 0
    else:
        return 1

df['Turbo'] = df['Engine volume'].apply(lambda x: TurboClass(x))
df['Engine volume'] = df['Engine volume'].apply(lambda x:convToVol(x))
print(df.head(5))
```

# Eliminating Models with less than 5 counts

```python
filtered_models = df.Model.value_counts().head(300)
def ModelElim(n):
    n = str(n).strip()
    if n in filtered_models:
        return n
    else:
        return 'other'
df['Model'] = df.Model.apply(lambda x: ModelElim(x))
df['Model'].value_counts()
```

# Dummies

```python
y = df['Price']
X = df.loc[:, ['Prod. year', 'Leather interior', 'Mileage', 'Cylinders', 'Airbags', 'Manufacturer', 'Model', 'Category', 'Fuel type', 'Engine volume', 'Gear box type', 'Drive wheels', 'Wheel', 'Color', 'Turbo']]
```

```python
#Getting dummies
car_cat = X.select_dtypes(include=['object'])

cars_dummies = pd.get_dummies(car_cat, drop_first=True)

X = X.drop(list(car_cat.columns), axis = 1)

X = pd.concat([X,cars_dummies], axis=1)
```

```python
cars_dummies.head()
```

# Feature Scaling


```python
from sklearn.preprocessing import scale
columns = X.columns
X = pd.DataFrame(scale(X))
X.columns = columns
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
l = LinearRegression()

l.fit(X_train, y_train)
y_pred = l.predict(X_test)

from sklearn.metrics import r2_score
print(r2_score(y_true=y_test, y_pred=y_pred))
```
